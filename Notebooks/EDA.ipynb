```python
import pandas as pd

# 1. Cargar los datos
car_data = pd.read_csv('../vehicles_us.csv')

# 2. Revisión inicial
print(car_data.info())
print(car_data.head())

# 3. Limpieza de datos
car_data['is_4wd'] = car_data['is_4wd'].fillna(0)

# Rellenar valores ausentes en 'model_year' y 'odometer' con la mediana para no afectar la distribución
car_data['model_year'] = car_data['model_year'].fillna(car_data['model_year'].median())
car_data['odometer'] = car_data['odometer'].fillna(car_data['odometer'].median())

# Eliminar filas con valores nulos en 'paint_color' o reemplazarlos por 'unknown'
car_data['paint_color'] = car_data['paint_color'].fillna('unknown')

# 4. Guardamos el dataset limpio para la app 
car_data.to_csv('vehicles_us_cleaned.csv', index=False)

print(car_data.info())
print(car_data.head())
```
